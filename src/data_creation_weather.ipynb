```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data = pd.read_csv('../data/1961678.csv')
#print(data.columns)
columns_we_want = ['STATION','NAME','DATE','AWND','PRCP','SNOW','TAVG','TMAX','TMIN']
boston_weather = data[data['NAME'] == 'BOSTON, MA US'][columns_we_want]
print(len(boston_weather))
boston_weather['Date/Time'] = pd.to_datetime(boston_weather['DATE'])
plt.scatter(np.array(boston_weather['Date/Time']),boston_weather['TAVG'].values)
plt.tight_layout()
plt.show()
plt.scatter(np.array(boston_weather['Date/Time']),boston_weather['PRCP'].values)
plt.tight_layout()
plt.show()
```

```python
export_csv = boston_weather.to_csv (r'../data/Processed_Data.csv', index = None, header = True)
```
